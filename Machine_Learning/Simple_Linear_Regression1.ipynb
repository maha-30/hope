```python
#import pandas for operation based for files
import pandas as pd
```

```python
#getting dataset
dataset=pd.read_csv("DataSet-SLAssignment.csv")
dataset
```

```python
#input
independent=dataset[["YearsOfExperience"]]
independent
```

```python
#salary input
dependent=dataset[["Salary"]]
dependent
```

```python
#all regression model will be imported from sklearn'
#this will be used to split the training set and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
y_test
```

```python
#y=mx+c
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#mass value
m_weight=regressor.coef_
m_weight
```

```python
#initial value or base
c_bias=regressor.intercept_
c_bias
```

```python
#get the prediction output based on test input
y_predict=regressor.predict(x_test)
y_predict
```

```python
#evaluation metrics
from sklearn.metrics import r2_score
```

```python
#if r2 score value near to 1 - good model else not good model
r_score=r2_score(y_test,y_predict)
r_score
```

```python
#create a file to save model
import pickle
filename="ML_Supervise_Regression_Linear.sav"
```

```python
#save the model into file
pickle.dump(regressor,open(filename,"wb"))
```

```python
#load the model
loaded_model=pickle.load(open(filename,"rb"))
```
